```python

import pandas as pd
import altair as alt

```

```python


# Load your data
df = pd.read_csv("eia_data_with_derived_variables.csv")

# Filter years >= 2000
df = df[df["Year"] >= 2000]
df = df[df["State"] != "US"]

# Compute average CO2 emissions per state per year
avg_df = (
    df.groupby(["State", "Year"], as_index=False)
      .agg({"CO2_Emissions": "mean"})
)
```

```python
# Find min and max per state
# Get index of min and max per state
idx_min = avg_df.groupby("State")["CO2_Emissions"].idxmin()
idx_max = avg_df.groupby("State")["CO2_Emissions"].idxmax()

# Select rows
extremes = pd.concat([
    avg_df.loc[idx_min],
    avg_df.loc[idx_max]
]).reset_index(drop=True)
```

```python
line = alt.Chart(avg_df).mark_line().encode(
    x="Year:O",
    y="CO2_Emissions:Q",
    color="State:N",
    tooltip=["State", "Year", "CO2_Emissions"]
)
```

```python
points = alt.Chart(extremes).mark_circle(size=100).encode(
    x="Year:O",
    y="CO2_Emissions:Q",
    color=alt.Color("State:N", legend=None),
    tooltip=[
        alt.Tooltip("State:N"),
        alt.Tooltip("Year:O"),
        alt.Tooltip("CO2_Emissions:Q"),
    ]
)
```

```python
chart = line + points
chart.interactive()
```
